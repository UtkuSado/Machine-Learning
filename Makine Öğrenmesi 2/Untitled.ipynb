```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
metre = [4.88, 7.75, 9.85, 12.05, 14.35, 16.75, 19.25, 21.85, 24.68, 27.62]
kilo = [43870, 26620, 20250, 15920, 12800, 10450, 8620, 7150, 5910, 4890]
```

```python
df_m = pd.DataFrame(metre)
df_m
df_k = pd.DataFrame(kilo)
df_k
df = pd.concat([df_m, df_k], axis= 1,)
df.columns = ['metre','kilo']
df
```

```python
import statsmodels.api as sm

y = df[["kilo"]]
X = df[['metre']]
```

```python
X = sm.add_constant(X)
X
```

```python
lm = sm.OLS(y, X)

model = lm.fit()

model.summary()
```
